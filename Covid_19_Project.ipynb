# Group by

```python
df_country=df.groupby(['Country'])[['Confirmed','Deaths', 'Recovered','Active']].sum().reset_index()
```

```python
df_country
```

```python
confirmed_cases=df.groupby(['Date'])['Confirmed'].sum().reset_index()
```

```python
confirmed_cases
```

```python
plt.figure(figsize=(12,10))
sns.lineplot(data=confirmed_cases, x='Date', y='Confirmed')
plt.title("confired cases over the time")
plt.xlabel('Date')
plt.ylabel('Confirmed cases')
plt.show()
```

```python
Active_cases=df.groupby(['Date'])['Active'].sum().reset_index()
```

```python
Active_cases
```

```python
Recovered_cases=df.groupby(['Date'])['Recovered'].sum().reset_index()
```

```python
Recovered_cases
```

```python
Deaths_cases=df.groupby(['Date'])['Deaths'].sum().reset_index()
```

```python
Deaths_cases
```

```python
from prophet import Prophet
```

```python
model=Prophet()
```

```python
# Prophet columns: ds (Date), y--> ds, y
```

```python
confirmed_cases.rename(columns={'Date':'ds', 'Confirmed':"y"},inplace=True)
confirmed_cases
```

```python
model.fit(confirmed_cases)
```

```python
future=model.make_future_dataframe(periods=7)
future
```

```python
#10
```

```python
future
```

```python
forcasted=model.predict(future)
```

```python
forcasted[['ds','yhat_lower', 'yhat_upper', 'trend_lower']].tail(7)
```

```python
model.plot(forcasted)
plt.show()
```
